```python
import pandas as pd
import altair as alt

alt.renderers.enable('default')
alt.data_transformers.disable_max_rows()
```

```python
url = "https://github.com/UIUC-iSchool-DataViz/is445_data/raw/main/licenses_fall2022.csv"
df = pd.read_csv(url)

df.head()
```

```python
df.columns
```

```python
# Convert expiration date to datetime
df['Expiration Date'] = pd.to_datetime(
    df['Expiration Date'], errors='coerce'
)

# Extract year
df['Expiration Year'] = df['Expiration Date'].dt.year

# Drop missing values
df_clean = df.dropna(subset=['License Type', 'Expiration Year'])
```

```python
top_categories = df_clean['License Type'].value_counts().head(10).index
df_top = df_clean[df_clean['License Type'].isin(top_categories)]
```

```python
license_options = sorted(df_top['License Type'].unique().tolist())
```

```python
plot1 = alt.Chart(df_top).mark_bar().encode(
    x=alt.X('count():Q', title='Number of Licenses'),
    y=alt.Y('License Type:N', sort='-x', title='License Type'),
    color=alt.Color('License Type:N', legend=None),
    tooltip=['License Type:N', 'count():Q']
).properties(
    width=600,
    height=400,
    title='Number of Records by License Type'
)

plot1
```

```python
input_dropdown = alt.binding_select(options=license_options, name='License Type: ')
selection = alt.selection_point(fields=['License Type'], bind=input_dropdown)
```

```python
plot2 = alt.Chart(df_top).mark_circle(size=80).encode(
    x=alt.X('Expiration Year:Q', title='Expiration Year'),
    y=alt.Y('License Type:N', title='License Type'),
    color=alt.Color('License Type:N', legend=None),
    tooltip=['License Type:N', 'Expiration Year:Q']
).add_params(
    selection
).transform_filter(
    selection
).properties(
    width=600,
    height=400,
    title='Expiration Year by Selected License Type'
)

plot2
```

```python
plot1
plot2
```

```python
plot1.save('plot1.html')
plot2.save('plot2.html')
```
